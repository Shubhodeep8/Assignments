```python
import pandas as pd 
import datetime as dt

sales_data=pd.read_excel("SalesData.xlsx")
sales_data.head()
```

```python
#1. Find the least amount sale that was done for each item. 

# Q1
least_amount=sales_data.groupby("Item")["Sale_amt"].min().reset_index()
least_amount
```

```python
#2. Compute the total sales for each year and region across all items

sales_data["OrderDate"]=pd.to_datetime(sales_data["OrderDate"])
sales_data["Year"]=sales_data["OrderDate"].dt.year
total_saless=sales_data.groupby(["Year","Region"])["Sale_amt"].sum().reset_index()
total_saless
```

```python
#3. Create new column 'days_diff' with number of days difference between reference date passed and each order date

reference_date = dt.datetime.today()
sales_data["days_diff"]=(reference_date-sales_data["OrderDate"]).dt.days
sales_data[["OrderDate","days_diff"]].head()
```

```python
#4. Create a dataframe with two columns: 'manager', 'list_of_salesmen'. Column 'manager' will contain the unique managers present and column 'list_of_salesmen' will contain an array of all salesmen under each manager. 
manager_salesman=sales_data.groupby("Manager")["SalesMan"].unique().reset_index()
manager_salesman.columns=["manager","list_of_salesman"]
manager_salesman
```

```python
#5. For all regions find number of salesman and total sales. Return as a dataframe with three columns - Region, salesmen_count and total_sales

region=sales_data.groupby("Region")

salesman_count=region["SalesMan"].nunique()

total_sales=region["Sale_amt"].sum()

data={
    "Region" : salesman_count.index,
    "Salesman_count" : salesman_count.values,
    "Total_Sales":total_sales.values
}
pd.DataFrame(data)
```

```python
#6. Create a dataframe with total sales as percentage for each manager. Dataframe to contain manager and percent_sales 

totall_sales=sales_data["Sale_amt"].sum()

manager_sales=sales_data.groupby("Manager")["Sale_amt"].sum()

percent_sales=(manager_sales/totall_sales)*100

data_sales = {
    "Manager" : percent_sales.index,
    "Percent_Sales" : percent_sales.values 
}
pd.DataFrame(data_sales)
```

```python
imdb_data = pd.read_csv("imdb.csv", on_bad_lines="skip")
imdb_data.head()

```

```python
#8. Return titles of movies with shortest and longest run time 


shortest_movie=imdb_data.loc[imdb_data["duration"].idxmin(),"title"]
longest_movie=imdb_data.loc[imdb_data["duration"].idxmax(),"title"]

print("Shortest movie : ",shortest_movie)
print("Longest movie : ",longest_movie)
```

```python
#9. Sort the data frame by in the order of when they were released and have higher ratings, Hint : release_date (earliest) and Imdb rating(highest to lowest) 


imdb_sorted=imdb_data.sort_values(by=["year","imdbRating"],ascending=[True,False])
imdb_sorted.head(10)
```

```python
diamond_data=pd.read_csv("diamonds.csv")
diamond_data
```

```python
#11. Count the duplicate rows of diamonds DataFrame. 

diamond_duplicate=diamond_data.duplicated().sum()
print("Number of Duplicate Rows: ",diamond_duplicate)
```

```python
#12. Drop rows in case of missing values in carat and cut columns.

missing_values = diamond_data.dropna(subset=["carat","cut"])
missing_values
```

```python
#13. Subset the dataframe with only numeric columns.

numeric_values=diamond_data.select_dtypes(include='number')
numeric_values
```

```python
#14. Compute volume as (xyz) when depth is greater than 60. In case of depth less than 60 default volume to 8

diamond_data["volume"] = diamond_data.apply(lambda row: row["x"] * row["y"] * row["z"] if row["depth"] > 60 else 8, axis=1)
diamond_data[["x", "y", "z", "depth", "volume"]].head()

```

```python
#15. Impute missing price values with mean.

mean = diamond_data["price"].mean()
diamond_data["price"].fillna(mean, inplace=True)
diamond_data
```
